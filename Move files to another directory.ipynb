## Copy files to another directory under criteria

```python
# Copy files to another directory

#iterate dataframe (go through each value)
#In Pandas Dataframe we can iterate an element in two ways: #Iterating over rows & #Iterating over columns

#source = os.listdir("/tmp/")
#destination = "/tmp/newfolder/"
path_src = "C:/1_TAM AUFINIA/EEI/EEI/1_Original_20190730/"
path_dtr = "C:/1_TAM AUFINIA/EEI/EEI/List DTR/"
path_tsh = "C:/1_TAM AUFINIA/EEI/EEI/List Timesheet/"
path_pay = "C:/1_TAM AUFINIA/EEI/EEI/List Payroll/"
path_test = "C:/1_TAM AUFINIA/EEI/EEI/Z_Test Moving file"

source = os.listdir(path_src) #print all the files in a directory path
destination_dtr = path_dtr
destination_tsh = path_tsh
destination_pay = path_pay
destination_test = path_test

str_tsh = ['TIMESHEET']
str_dtr = ['DTR']
str_both = ['TIMESHEET','DTR']

#i = 0
#https://treyhunner.com/2016/04/how-to-loop-with-indexes-in-python/
for i in range(len(source)):
    if not any(str in source[i] for str in str_both):
        #print('result 1: ' + source[i])
        shutil.copy(path_src + source[i], destination_pay)
    elif any(str in source[i] for str in str_tsh):
        #print('result 2: ' + source[i])
        shutil.copy(path_src + source[i], destination_tsh)
    elif any(str in source[i] for str in str_dtr):
        #print('result 3: ' + source[i])
        shutil.copy(path_src + source[i], destination_dtr)
```
