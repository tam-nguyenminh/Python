# Import Data

```python
crash_df = sns.load_dataset('car_crashes')
crash_df.head(10)
```

# Distribution Plot

```python
sns.distplot(crash_df['not_distracted'],kde= True, bins = 25)
```

# Join Plot

```python
sns.jointplot(x='speeding', y='alcohol', data= crash_df, kind='reg')
```

```python
sns.jointplot(x='speeding', y='alcohol', data= crash_df, kind='kde')
```

```python
sns.jointplot(x='speeding', y='alcohol', data= crash_df, kind='hex')
```

# KDE plot

```python
sns.kdeplot(crash_df['alcohol'])
```

# Pair Plot

```python
sns.pairplot(crash_df)
```

```python
tips_df = sns.load_dataset('tips')
tips_df
```

```python
sns.pairplot(tips_df, hue='sex', palette='Blues')
```

# Rug Plot

```python
sns.rugplot(tips_df['tip'])
# shows data points as sticks to see its density, similar to histogram
```

# Styling

```python
sns.set_style('ticks') #darkgrid, dark, ticks, white
plt.figure(figsize=(8,4))

sns.set_context('paper',font_scale=1.4) #paper, talk, poster

sns.jointplot(x='speeding', y='alcohol', data=crash_df, kind='reg')

# left, bottom axis line
# sns.despine(left=True, bottom=True)
```

# Catergorical Plot

## Bar plot

```python
sns.barplot(x='sex',y='total_bill', data = tips_df)
# based on the mean by default
```

```python
sns.barplot(x='sex',y='total_bill', data = tips_df, estimator=np.std)
# can change the aggregation by estimator
# median, np.median, np.std, np.var, np.cov or even customize
```

## Count plot

```python
sns.countplot(x='sex', data=tips_df)
```
