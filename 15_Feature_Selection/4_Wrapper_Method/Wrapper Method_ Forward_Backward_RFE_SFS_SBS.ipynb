# Forward Selection

```python
# X, y = df, breast_cancer.target.reshape(-1,1)
X, y = breast_cancer.data, breast_cancer.target.reshape(-1,1)
feature_names = breast_cancer.feature_names
```

```python
print(X.shape, y.shape)
```

```python
X[:,1].reshape(-1,1)
```

```python
scores = []
for i in range(X.shape[1]):
    # print(X.columns[i])
    X_sel = X[:,i].reshape(-1,1)
    X_train, X_test, y_train, y_test = train_test_split(X_sel, y, test_size=0.3, random_state=42)
    model = LogisticRegression()
    model.fit(X_train, y_train)
    score = model.score(X_test, y_test)
    scores.append(score)
scores
```

```python
forward_select_res = pd.DataFrame(zip(breast_cancer.feature_names, scores), columns = ['feature', 'score']).sort_values('score', ascending=False)
forward_select_res
```

# Recursive Feature Elimination RFE
Feature ranking with recursive feature elimination.

Given an external estimator that assigns weights to features (e.g., the coefficients of a linear model), \
the goal of recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features. \
First, the estimator is trained on the initial set of features and the importance of each feature is obtained \
either through any specific attribute or callable. \
Then, the least important features are pruned from current set of features. \
That procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached.

```python
from sklearn.feature_selection import RFECV, RFE
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
X, y = df, breast_cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
rfecv = RFECV(LogisticRegression(), min_features_to_select=5,verbose=5, cv=5, scoring='accuracy')
# rfe = RFE(LogisticRegression(), n_features_to_select=7, verbose=5, step=1)
rfecv_fit = rfecv.fit(X_train, y_train)
```

# Sequential Feature Selection
Sequential Forward Selection (SFS)\
Sequential Backward Selection (SBS)

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
X, y = df, breast_cancer.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=12)
```

```python
sfs = SFS(RandomForestClassifier(), 
          k_features=5, #number of features to select
          forward = True, #rolling forward selection
          floating=False, # No floating selection
          scoring='accuracy',
          verbose=5,
          cv=5
          ).fit(X_train, y_train)
sfs
```
